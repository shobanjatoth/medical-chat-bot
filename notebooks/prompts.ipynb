```python
import os
from langchain_core.runnables import RunnableLambda
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableSequence
from langchain_google_genai import ChatGoogleGenerativeAI

GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")

llm = ChatGoogleGenerativeAI(
    api_key=GOOGLE_API_KEY,
    model="gemini-2.5-flash"
)
prompt=PromptTemplate(template="explain the {topic} in 100 to 300 words",input_variables=["topic"])
parser=StrOutputParser()
def upper(text):
    return text.upper()

uppercase_output = RunnableLambda(upper)

chain = prompt | llm | parser | uppercase_output

result = chain.invoke({"topic": "rohit"})

print(result)
def upper(text):
    return text.upper()
uppercase_output=RunnableSequence(upper)
chain=prompt|llm|parser|uppercase_output
chain.invoke({"topic":"rohit"})
```
